```python
geoviz = leafmap.Map()
url = 'https://storage.googleapis.com/pdd-stac/disasters/hurricane-harvey/0831/2170831_172754_101c_3b_Visual.tif'
geoviz.add_cog_layer(url, name = "Planet-Harvey-2017")
geoviz
```

```python
m = leafmap.Map()
url = 'https://opendata.digitalglobe.com/events/california-fire-2020/pre-event/2018-02-16/pine-gulch-fire20/1030010076004E00.tif'
m.add_cog_layer(url, name="Fire (pre-event)")
m
```

```python
m = leafmap.Map(center=[0, 0], zoom=2)
in_geojson = 'https://raw.githubusercontent.com/giswqs/leafmap/master/examples/data/cable-geo.geojson'
m.add_geojson(in_geojson, layer_name="Cable lines")
m
```
